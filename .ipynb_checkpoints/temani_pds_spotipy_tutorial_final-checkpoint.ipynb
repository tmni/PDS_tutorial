## Bieber vs Timberlake

Who's the more popular Justin? These are the core questions that data science can answer. WE can use the popularity feature as well as the search rank order to determine this.

```python
def popularity(search_term):
    results = sp.search(q=search_term, limit=5, type="artist") #retrieve search results to rank order
    d = {"rank":[], "artist":[], "popularity":[]} 
    
    for i, t in enumerate(results['artists']['items']):
        r = i+1
        d["rank"].append(r)
        a =  t['name']
        d["artist"].append(a)
        p = t['popularity']
        d["popularity"].append(p)
    df = pd.DataFrame(data=d)
    df = df.set_index('rank')
    return df
pop = popularity("Justin") #change the name to check it out for yourself!
pop
```

Let's plot this to see how close Bieber and Timberlake are

```python
def plotpop(df):
    
    b = sns.barplot(x=pop.artist, y=pop.popularity, palette="BuGn_r")
    plt.xticks(rotation=90)
    return b
plotpop(pop)
```

Truly Tragic. We can see that Jusin Timberlake is not only outranked by Justin Bieber, but he's not even the second in popularity! How times have changed...feel free to change the input of the function to see how other names compare. (Selena or Selena Gomez, Michael Jackson or Michael Bublé, etc)

### Compare Decades
Here's a method that takes in playlists and returns a list of dataframes with each songs audio features. I inputted the best of 70s, 80s, 90s, 00s from Spotify and the top hits of today. You can get the URIs from going to the playlist in the app and  clicking on the three dots.

```python
def compareDecades(l):
    
    dfs = [] #hold list of playlist dfs
    for uri in l: 
        username = uri.split(':')[2]
        playlist_id = uri.split(':')[3]
        results = sp.user_playlist(username, playlist_id)
        tids = []
        for i, t in enumerate(results['tracks']['items']):
            wow = t['track']
            tids.append(wow['uri'])
        start = time.time()
        features = sp.audio_features(tids) #retrieve audio features
        delta = time.time() - start #calculate how long it takes
        df = pd.DataFrame(features)  
        s = df.mean()
        s = s.reset_index()
        s = s.set_index('index')
        dfs.append(df)
        print ("features retrieved in %.2f seconds" % (delta,))
    return dfs


```

```python
#decade playlists
seventies = "spotify:user:playlist:37i9dQZF1DWTJ7xPn4vNaz"
eighties = "spotify:user:playlist:37i9dQZF1DX4UtSsGT1Sbe"
nineties = "spotify:user:playlist:37i9dQZF1DXbTxeAdrVG2l"
thousands = "spotify:user:playlist:37i9dQZF1DX4o1oenSJRJd"
now = 'spotify:user:playlist:37i9dQZF1DXcBWIGoYBM5M'

uris = [seventies, eighties, nineties, thousands, now]
decades = compareDecades(uris)
```

Here's what a dataframe of one of the decades looks like:

```python
def plotDecades(dflist):
    seventiesdf,eightiesdf, ninetiesdf, thousandsdf, nowdf = dflist[0], dflist[1], dflist[2], dflist[3], dflist[4]
    concatenated = pd.concat([seventiesdf.assign(dataset='70s'), 
                          eightiesdf.assign(dataset='80s'), 
                          ninetiesdf.assign(dataset='90s'),
                          thousandsdf.assign(dataset='00s'),
                          nowdf.assign(dataset='present')])
    v = sns.catplot(x="dataset", y="valence", kind="box",  data=concatenated, palette="BuGn_r")
    vfriend = sns.swarmplot(x="dataset", y="valence", color="green", size=3, data=concatenated)
    
    e = sns.catplot(x="dataset", y="energy", kind="box",  data=concatenated, palette="BuGn_r")
    efriend = sns.swarmplot(x="dataset", y="energy", color="green", size=3, data=concatenated)
    
    d =sns.catplot(x="dataset", y="danceability", kind="box",  data=concatenated, palette="BuGn_r")
    dfriend =sns.swarmplot(x="dataset", y="danceability", color="green", size=3, data=concatenated)
    
    l =sns.catplot(x="dataset", y="loudness", kind="box",  data=concatenated, palette="BuGn_r")
    lfriend =sns.swarmplot(x="dataset", y="loudness", color="green", size=3, data=concatenated)

plotDecades(decades)
```

## Song Prediction

Spotify uses really intense machine learning algorithms to provide song predictions to the user. While we can't fully replicate that, let's make a basic training set and use the sklearn library to make predictions about whether you'll like a song or not by training on Spotify playlists that you like and dislike! This section of the tutorial was inspired by [Behren's article](https://towardsdatascience.com/making-your-own-discover-weekly-f1ac7546fedb) and altered to fit our needs. 

First, find 2 playlists. One with music you like, and one with music you dislike. We need thousands of songs in order to receive more accurate results, so I'm borrowing Behren's premade playlists for demonstration purposes. We then parse the playlists to return their IDs and Audio Features. See the code below:

```python
def getIds(user, playlist): 
    playlist = sp.user_playlist(user, playlist) 
    tracks = playlist["tracks"]
    songs = tracks["items"] 
    while tracks['next']: #get all the tracks
        tracks = sp.next(tracks)
        for item in tracks["items"]:
            songs.append(item)
    ids = [] 
    for i in range(len(songs)):
        ids.append(songs[i]['track']['id']) # add ids to a list
    return ids
        
def getFeatures(good_ids, bad_ids):   
    features = [] #store all song features
    for i in range(0,len(good_ids), 50): #skip by 50 because spotipy's limit is 50
        audio_features = sp.audio_features(good_ids[i:i+50])
        for track in audio_features:
            features.append(track)
            features[-1]['target'] = 1 
    for i in range(0,len(bad_ids),50): #separate playlists to avoid double loop
        audio_features = sp.audio_features(good_ids[i:i+50])
        for track in audio_features:
            features.append(track)
            features[-1]['target'] = 0
    return features #return list of songs and their features
```

Since there's 4000+ songs, it will take about 30 sec to run. Feel free to grab a coffee while you wait. ☕

```python
good_ids= getIds("1287242681", "5OdH7PmotfAO7qDGxKdw3J")
bad_ids = getIds("1287242681", "3ySDAXYGUwRrp8C4ejIm9m")
data = getFeatures(good_ids, bad_ids)
print("Total Songs in Data Set: ", len(data))
```

Here you can see that we have successfully inputted the songs for the playlist and parsed all their features. Let's turn it into a dataframe for easier visibility.

```python
training_data = pd.DataFrame(data)
training_data
```

You might remember from class, but when using data for training a model, we need to split it into training and testing. Slearn's train_test_split does this. Here we are splitting the model 80 and 20, respectively. We then define what features we are going to be looking at for relevance in determining what a song is "worth."

```python
train, test = train_test_split(training_data, test_size = 0.20)
print("Training size: {}, Test size: {}".format(len(train),len(test)))

features = ["danceability", "loudness", "valence", "energy", "instrumentalness", "acousticness", "key", "speechiness","duration_ms"]
```

Now for the fun part, let's use sklearn's SVC method to fit our data. The kernel selects the type of hyperplane used to separate the data. C is the penalty parameter of the error term. It controls the trade off between smooth decision boundary and classifying the training points correctly. We can see how accurate this model is as well, which means how accurate our classification is. This also takes a bit of time so drink some more coffee __#midtermszn__). ☕ ☕ ☕ 

```python
#Train train data
x_train = train[features] 
y_train = train["target"]
#Test test data
x_test = test[features]
y_test = test["target"]

svc_lin = SVC(kernel="linear", C=0.025) #set parameters for model
svc_lin.fit(x_train, y_train) #fit with data given
svc_pred = svc_lin.predict(x_test) #compare with the test
print(svc_lin)

score = accuracy_score(y_test, svc_pred) * 100
print("Accuracy using svc linear: ", round(score, 1), "%")
```

It's a bit low, but this can be improved with adding more data (larger playlists) and playing around with the numbers. Feel free to try it out! 

Now, lets use our SVC model to predict whether we'll like songs in a playlist. First, lets put our playlist to test in a desired format:

```python
def playlistToTest(username, playlist):
    playlist = sp.user_playlist(username, playlist) #retrieve playlist from api
    newPlaylist_tracks = playlist["tracks"]
    newPlaylist_songs = newPlaylist_tracks["items"] 
    while newPlaylist_tracks['next']: #go through songs and append to songlist
        newPlaylist_tracks = sp.next(newPlaylist_tracks)
        for song in newPlaylist_tracks["items"]:
            newPlaylist_songs.append(song)

    newPlaylist_song_ids = [] 
    for i in range(len(newPlaylist_songs)):
        newPlaylist_song_ids.append(newPlaylist_songs[i]['track']['id'])

    newPlaylist_features = []
    j = 0
    for i in range(0,len(newPlaylist_song_ids),50):
        audio_features = sp.audio_features(newPlaylist_song_ids[i:i+50])
        for track in audio_features:
            track['song_title'] = newPlaylist_songs[j]['track']['name']
            track['artist'] = newPlaylist_songs[j]['track']['artists'][0]['name']
            j= j + 1
            newPlaylist_features.append(track) #return playlist and its features
    playlist_features = pd.DataFrame(newPlaylist_features)
    return playlist_features

```

This function below will take in that cleaned up playlist and use the model we created early to predict whether this song will be liked and return them if so. 

```python
def WillILikeThis(df): #function where you input playlist and see which songs you'll like
    pred = svc_lin.predict(df[features])
    likedSongs = 0
    i = 0
    for prediction in pred:
        print ("Song: " + df["song_title"][i] + ", By: "+ df["artist"][i])
        if(prediction == 1):
            likedSongs= likedSongs + 1
        i = i +1
    print("found ", i, "songs you may like in this playlist")
```

```python
df = playlistToTest('spotify', '37i9dQZF1DXcBWIGoYBM5M')
WillILikeThis(df)
```
